```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import pandas as pd
pd.options.mode.chained_assignment = None 

import scipy.stats as stats
```

```python
data = pd.read_csv("Clean_data")

data.head()
```

```python
data = data[["LotFrontage","LotArea","GrLivArea","SalePrice"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, # predictors
    data['SalePrice'],  
    test_size=0.3,  
    random_state=42)
```

```python
#function for plotting

def diag_plots(df, variable):
    
    """
    function to plot a histogram and a Q-Q plot
    side by side, for a certain variable
    """
    
    
    plt.figure(figsize=(15,5))
    plt.subplot(1, 2, 1)
    df[variable].hist(bins=50)
    plt.title(var)

    plt.subplot(1, 2, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.title(variable)

    plt.show()
```

```python
#creating the list of continious variable 
num_var = ["LotFrontage","LotArea","GrLivArea"]
```

```python
#creating the plots
for var in num_var:
    diag_plots(X_train,var)
```

```python
#applying the transformation
for var in num_var:
    X_train[var+"_exp"] = data[var]**(1/4)
```

```python
#list of transformed variables
transformed = ["LotFrontage_exp","LotArea_exp","GrLivArea_exp"]
```

```python
#plotting transformed variable
for var in transformed:
    diag_plots(X_train,var)
```
