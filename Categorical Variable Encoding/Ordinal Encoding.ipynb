# Ordinal Encoding Using Pandas

```python
#copying the data 
data=X_train.copy()
```

```python
data.head()
```

```python
cat_var = ["Alley","LotShape","GarageCond"]
```

```python
#function that gives ordinal values fo the data frame
def ordinal_mapper(data, var):
    
    """
    It takes data and variable and transform the categorical to numerical
    """
    
    #mapper
    ordinal_map = {
    k:i for i,k in enumerate(data[var].unique(),0)
    }
    
    #applying the map
    data[var]=data[var].map(ordinal_map)
```

```python
for var in cat_var:
    ordinal_mapper(data,var)
```

# Ordinal Encoding Using Sci-kit Learn

```python
#library to use
from sklearn.preprocessing import LabelEncoder
```

```python
#copying data
df = X_train.copy()
```

```python
df.head()
```

```python
#Ordinal encoding function

def encoder(data,var):
    """
    this function takes data and variabe and change it from categorical to numerical
    """
    
    encoder = LabelEncoder()
    encoder.fit(data[var])
    
    data[var] = encoder.transform(data[var])
```

```python
cat_var = ["Alley","LotShape","GarageCond"]
```

```python
for var in cat_var:
    encoder(df,var)
```
